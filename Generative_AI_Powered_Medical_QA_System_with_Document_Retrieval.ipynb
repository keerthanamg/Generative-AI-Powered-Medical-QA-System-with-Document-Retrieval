```python
from langchain_community.document_loaders import PyPDFDirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain_community.embeddings import SentenceTransformerEmbeddings
from langchain.vectorstores import Chroma
from langchain_community.llms import LlamaCpp
from langchain.chains import RetrievalQA, LLMChain
```

```python
loader = PyPDFDirectoryLoader("/content/drive/MyDrive/BioMistral")
docs = loader.load()
```

```python
len(docs)
```

```python
docs[5]
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)
chunks = text_splitter.split_documents(docs)
```

```python
len(chunks)
```

```python
chunks[5]
```

```python
import os
```

```python
os.environ['HUGGINGFACEHUB_API_TOKEN'] = "your huggingfacehub api token"
```

```python
embeddings = SentenceTransformerEmbeddings(model_name="NeuML/pubmedbert-base-embeddings")
```

```python
VectorStore = Chroma.from_documents(chunks, embeddings)
```

```python
query = "Who is at risk of heart disease?"
search_results= VectorStore.similarity_search(query)
```

```python
search_results
```

```python
retriver = VectorStore.as_retriever(search_kwargs={'k':5})
```

```python
retriver.get_relevant_documents(query)
```

```python
llm = LlamaCpp(
    model_path = "/content/drive/MyDrive/BioMistral/ggml-model-Q4_K_M.gguf",
    temperature = 0.2,
    max_tokens = 2048,
    top_p = 1
)
```

```python
template = """
<!context!>
You are a Medical Assistance that follows the instructions and generate the accurate response based on the query and the context provided. Please be truthful and give direct answers.
</s>
<!user>
{query}
</s>
<!assistant!>
"""
```

```python
from langchain.schema.runnable import RunnablePassthrough
from langchain.schema.output_parser import StrOutputParser
from langchain.prompts import ChatPromptTemplate
```

```python
prompt = ChatPromptTemplate.from_template(template)
```

```python
rag_chain = (
    {"context": retriver, "query": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
```

```python
response = rag_chain.invoke(query)
```

```python
response
```

```python
import sys
while True:
  user_input = input(f"Input query:")
  if user_input == 'exit':
    print("Exiting...")
    sys.exit()
  if user_input == "":
    continue
  result = rag_chain.invoke(user_input)
  print("Answer:", result)
```
